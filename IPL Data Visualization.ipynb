```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
ipl_data = pd.read_csv('Cricket_data.csv')
```

```python
ipl_data.dropna(inplace=True)
```

```python
ipl_data['description'] = ipl_data['description'].str.lower()
```

```python
ipl_data['home_team'] = ipl_data['home_team'].astype('category')
ipl_data['away_team'] = ipl_data['away_team'].astype('category')
```

```python
ipl_data['match_type'] = ipl_data['description'].str.extract(r'(\w+) \(N\)')
```

```python
ipl_data['season'] = ipl_data['season'].astype(int)
```

```python
ipl_data.dtypes
```

```python
ipl_data['start_date'] = pd.to_datetime(ipl_data['start_date'])
ipl_data['end_date'] = pd.to_datetime(ipl_data['end_date'])
ipl_data['toss_won'] = ipl_data['toss_won'].astype('category')
ipl_data['decision'] = ipl_data['decision'].astype('category')
ipl_data['1st_inning_score'] = pd.to_numeric(ipl_data['1st_inning_score'], errors='coerce')
ipl_data['2nd_inning_score'] = pd.to_numeric(ipl_data['2nd_inning_score'], errors='coerce')
ipl_data['winner'] = ipl_data['winner'].astype('category')
ipl_data['result'] = ipl_data['result'].astype('category')
ipl_data['home_overs'] = pd.to_numeric(ipl_data['home_overs'], errors='coerce')
ipl_data['home_runs'] = pd.to_numeric(ipl_data['home_runs'], errors='coerce')
ipl_data['home_wickets'] = pd.to_numeric(ipl_data['home_wickets'], errors='coerce')
ipl_data['home_boundaries'] = pd.to_numeric(ipl_data['home_boundaries'], errors='coerce')
ipl_data['away_overs'] = pd.to_numeric(ipl_data['away_overs'], errors='coerce')
ipl_data['away_runs'] = pd.to_numeric(ipl_data['away_runs'], errors='coerce')
ipl_data['away_wickets'] = pd.to_numeric(ipl_data['away_wickets'], errors='coerce')
ipl_data['away_boundaries'] = pd.to_numeric(ipl_data['away_boundaries'], errors='coerce')
ipl_data['match_days'] = pd.to_numeric(ipl_data['match_days'], errors='coerce')
ipl_data['umpire1'] = ipl_data['umpire1'].astype('category')
ipl_data['umpire2'] = ipl_data['umpire2'].astype('category')
ipl_data['tv_umpire'] = ipl_data['tv_umpire'].astype('category')
ipl_data['referee'] = ipl_data['referee'].astype('category')
ipl_data['reserve_umpire'] = ipl_data['reserve_umpire'].astype('category')
ipl_data['match_type'] = ipl_data['match_type'].astype('category')
```

```python
ipl_data.dtypes
```

```python
team_performance = ipl_data.groupby('winner').agg({
    'id': 'count',                    # Total matches played
    'decision': lambda x: (x == 'BAT FIRST').sum()
    
})

team_performance.columns = ['Total Matches', 'Matches Won Batting First']
team_performance['Win Percentage'] = (team_performance['Matches Won Batting First'] / team_performance['Total Matches']) * 100
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='season', data=ipl_data, palette='viridis')
plt.title('Distribution of Matches Across Seasons', fontsize=16)
plt.xlabel('Season', fontsize=14)
plt.ylabel('Number of Matches', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
winner_counts = ipl_data['winner'].value_counts().sort_values(ascending=False)
winner_plot = sns.barplot(x=winner_counts.index, y=winner_counts.values, palette='coolwarm')
plt.title('Distribution of Match Outcomes', fontsize=16)
plt.xlabel('Winner', fontsize=14)
plt.ylabel('Number of Matches Won', fontsize=14)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)
for p in winner_plot.patches:
    winner_plot.annotate(format(p.get_height(), '.0f'), 
                         (p.get_x() + p.get_width() / 2., p.get_height()), 
                         ha = 'center', va = 'center', 
                         xytext = (0, 10), 
                         textcoords = 'offset points', fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
corr = ipl_data.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, annot=True, fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='winner', data=ipl_data, palette='viridis', order=ipl_data['winner'].value_counts().index)
plt.title('Total Wins by Team')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(x='season', hue='winner', data=ipl_data, palette='muted')
plt.title('Wins by Team Across Seasons')
plt.xlabel('Season')
plt.ylabel('Number of Wins')
plt.legend(title='Team', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a new DataFrame to calculate wins by team in home matches
home_wins = ipl_data[ipl_data['home_team'] == ipl_data['winner']].groupby('home_team').size().reset_index(name='home_wins')

# Create a new DataFrame to calculate wins by team in away matches
away_wins = ipl_data[ipl_data['away_team'] == ipl_data['winner']].groupby('away_team').size().reset_index(name='away_wins')

# Merge the home and away wins DataFrames
total_wins = home_wins.merge(away_wins, left_on='home_team', right_on='away_team', how='outer')

# Rename columns for clarity
total_wins.rename(columns={'home_team': 'team', 'home_wins': 'home_wins', 'away_wins': 'away_wins'}, inplace=True)

# Fill NaN values with 0
total_wins.fillna(0, inplace=True)

# Calculate total wins by summing home and away wins
total_wins['total_wins'] = total_wins['home_wins'] + total_wins['away_wins']

# Sort the DataFrame by total wins
total_wins.sort_values(by='total_wins', ascending=False, inplace=True)

# Plot the bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='total_wins', y='team', data=total_wins, palette='muted')
plt.title('Total Wins by Team (Home and Away Matches)')
plt.xlabel('Number of Wins')
plt.ylabel('Team')
plt.show()
```

```python
# Extracting relevant columns
match_results = ipl_data[['season', 'winner']]

# Grouping the data by season and winner and counting the number of matches won by each team in each season
season_wise_winners = match_results.groupby(['season', 'winner']).size().unstack(fill_value=0)

# Plotting the distribution of match outcomes over the seasons using a stacked bar plot
season_wise_winners.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Distribution of Match Outcomes Over Seasons')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.legend(title='Team')
plt.show()
```

```python
# Calculate total wins for each team across all seasons
total_wins = match_results['winner'].value_counts()

colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'cyan', 'magenta']

# Plotting the total wins for each team with colorful bars
plt.figure(figsize=(12, 6))
total_wins.plot(kind='bar', color=colors)
plt.title('Total Wins by Each Team Across All Seasons')
plt.xlabel('Team')
plt.ylabel('Total Wins')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Summing up the runs scored in the first innings by home team for each season
runs_by_season_home_first = ipl_data.groupby(['season', 'home_team'])['1st_inning_score'].sum().reset_index()

# Summing up the runs scored in the second innings by home team for each season
runs_by_season_home_second = ipl_data.groupby(['season', 'home_team'])['2nd_inning_score'].sum().reset_index()

# Summing up the runs scored in the first innings by away team for each season
runs_by_season_away_first = ipl_data.groupby(['season', 'away_team'])['1st_inning_score'].sum().reset_index()

# Summing up the runs scored in the second innings by away team for each season
runs_by_season_away_second = ipl_data.groupby(['season', 'away_team'])['2nd_inning_score'].sum().reset_index()

# Pivot the data to have seasons as rows and teams as columns for both home and away teams
runs_by_season_home_first_pivot = runs_by_season_home_first.pivot(index='season', columns='home_team', values='1st_inning_score')
runs_by_season_home_second_pivot = runs_by_season_home_second.pivot(index='season', columns='home_team', values='2nd_inning_score')
runs_by_season_away_first_pivot = runs_by_season_away_first.pivot(index='season', columns='away_team', values='1st_inning_score')
runs_by_season_away_second_pivot = runs_by_season_away_second.pivot(index='season', columns='away_team', values='2nd_inning_score')

# Plotting total runs scored by home teams over seasons
plt.figure(figsize=(12, 6))
runs_by_season_home_first_pivot.plot(kind='bar', stacked=True, cmap='tab20')
plt.title('Total Runs Distribution by Home Teams Over Seasons (1st Innings)')
plt.xlabel('Season')
plt.ylabel('Total Runs')
plt.legend(title='Home Team')
plt.xticks(rotation=45)
plt.show()

# Plotting total runs scored by away teams over seasons
plt.figure(figsize=(12, 6))
runs_by_season_away_first_pivot.plot(kind='bar', stacked=True, cmap='tab20')
plt.title('Total Runs Distribution by Away Teams Over Seasons (1st Innings)')
plt.xlabel('Season')
plt.ylabel('Total Runs')
plt.legend(title='Away Team')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting total runs scored by home teams in the second innings over seasons
plt.figure(figsize=(12, 6))
runs_by_season_home_second_pivot.plot(kind='bar', stacked=True, cmap='tab20')
plt.title('Total Runs Distribution by Home Teams Over Seasons (2nd Innings)')
plt.xlabel('Season')
plt.ylabel('Total Runs')
plt.legend(title='Home Team')
plt.xticks(rotation=45)
plt.show()

# Plotting total runs scored by away teams in the second innings over seasons
plt.figure(figsize=(12, 6))
runs_by_season_away_second_pivot.plot(kind='bar', stacked=True, cmap='tab20')
plt.title('Total Runs Distribution by Away Teams Over Seasons (2nd Innings)')
plt.xlabel('Season')
plt.ylabel('Total Runs')
plt.legend(title='Away Team')
plt.xticks(rotation=45)
plt.show()
```
